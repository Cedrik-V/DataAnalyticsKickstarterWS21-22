### Die Bearbeitung der Anforderungen 3 und 4 erfolgt in diesem Notebook

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
from os.path import dirname
import datetime as dt
import seaborn as sns
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.feature_selection import SelectFromModel
from sklearn.svm import LinearSVC
from sklearn.pipeline import Pipeline
from sklearn.svm import SVC
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.model_selection import cross_val_predict
from xgboost import XGBClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.feature_selection import VarianceThreshold


notebook_path = os.path.abspath("6. b) Klassifikation mit verschiedenen Lernern.ipynb")
csv_path_features = os.path.join(os.path.dirname(dirname(notebook_path)), "data/ks-project-edited-klassifikation-features.csv")
csv_path_trueLabels = os.path.join(os.path.dirname(dirname(notebook_path)), "data/ks-project-edited-klassifikation-target.csv")
```

```python
df_features = pd.read_csv (csv_path_features, low_memory=False)
df_target = pd.read_csv (csv_path_trueLabels, low_memory=False)
```

```python
df_features.head()
```

```python
df_target.head()
```

Da unser Datensatz ist zu groß ist, um ihn vollständig für verschiedene Klassifizierer durchlaufen zu lassen. Deshalb erstellen wir ein Subset von unserem Dataframe mit 10000 Einträgen. (Zuvor haben wir die Ergebnisse von Subsets mit verschiedenen Größen verglichen)

```python
df_features["target"] = df_target["successful"]
df_features.head()
```

```python
df_sample = df_features.sample(n=10000, random_state=42)
df_sample
```

```python
X_train, X_dev, y_train, y_dev = train_test_split(df_sample.iloc[: ,0:19], df_sample.iloc[: , -1], random_state=42)
```

```python
feature_selection_1=SelectFromModel(LinearSVC(penalty="l1", dual=False,max_iter=2000))
clf_list = []
# Initialiserung der verschiedenen Klassifizierer

clf1 = Pipeline([('feature_selection', feature_selection_1),
                     ('classifier', SVC())])
clf_list.append(clf1)

clf2 = Pipeline([('feature_selection', feature_selection_1),
                     ('classifier', RandomForestClassifier())])
clf_list.append(clf2)

clf3 = Pipeline([('feature_selection', feature_selection_1),
                     ('classifier', KNeighborsClassifier())])
clf_list.append(clf3)

clf4 = Pipeline([('feature_selection', feature_selection_1),
                     ('classifier', GaussianNB())])
clf_list.append(clf4)

clf5 = Pipeline([('feature_selection', feature_selection_1),
                     ('classifier', XGBClassifier(use_label_encoder =False))])
clf_list.append(clf5)
```

```python
# Setzen von verschiedenen Hyperparametern für die Klassifizierer, welche dann mit GridSearch getestet werden
search_spaces =[]
                
search_space1 = [
                {'feature_selection__threshold': [None, 'mean'], 'classifier__C': [0.001, 0.01, 0.1, 1, 10, 100], 'classifier__gamma': [0.001, 0.01, 0.1, 1, 10, 100], 'classifier__kernel': ['rbf']}
                ]
search_spaces.append(search_space1)

search_space2 = {'feature_selection__threshold': [None, 'mean'],
                 'classifier__n_estimators': [10, 100],
                 'classifier__min_samples_split': [2,4],
                 'classifier__max_features': ['auto','log2']
                }
search_spaces.append(search_space2)

search_space3 = {'feature_selection__threshold': [None, 'mean'],
                 'classifier__n_neighbors': [3, 7, 11],
                 'classifier__weights': ['uniform', 'distance'],
                 'classifier__metric': ['minkowski','euclidean', 'manhatten']
                }
search_spaces.append(search_space3)

search_space4 = {'feature_selection__threshold': [None, 'mean']
                }
search_spaces.append(search_space4)

search_space5 = {'feature_selection__threshold': [None, 'mean'],
                 'classifier__gamma': [0.001, 0.01, 0.1, 1, 10, 100],
                 'classifier__max_depth': [5,6,7],
                }
search_spaces.append(search_space5)
```

### Anforderung 3 Beschreiben des Vorgehens

Insgesamt haben wir fünf unterschiedliche Klassifikationsalgorithmen getestet, die Support Vector Machine, Random Forest, Nearest Neighbors, Naive Bayes und den eXtream Gradient Boosting (XGBoost) Klassifizierer. Für jeden dieser Klassifikationsalgorithmen haben wir eine Pipeline erstellt in der vor der Klassifizierung noch eine Feature-Auswahl mit der SelectFromModel Methode der wir als „estimator“ die LinearSVC übergeben haben. Durch diesen ersten Schritt in der Pipeline werden nur die Features zum Trainieren des Modells ausgewählt, die als wichtig genug eingeschätzt werden anhand ihrer Gewichtung durch den „estimator“. 
Zum Testen der verschiedenen Klassifikationsalgorithmen haben wir für jeden Klassifizierer verschiedene Werte gewählt für bestimmte Hyperparameter und unsere Datensatz in Trainings- und Testdaten unterteilt. Mithilfe von GridSearch haben wir dann alle Parameterkombinationen auf unseren Trainingsdaten getestet. Unser bester Lerner nach diesem Schritt war der XGBoost Klassifizierer mit einem durchschnittlich cross-validierten Score von 0,727.

```python
best_params = []
best_scores = []
for i in range(0,5):
#f"search_space{i}",
    print("---------------------------------------------------------------------------")
    grid_search = GridSearchCV(clf_list[i],param_grid=search_spaces[i], verbose=3)
    grid_search.fit(X_train, y_train)
    best_params.append(grid_search.best_estimator_)
    best_scores.append(grid_search.best_score_)
```

```python
# Ausgabe der besten Parameter für die Klassifizierer
print(best_params)
# Ausgabe der besten Scores
print(best_scores)
```

### Anforderung 3 Performanz auf den Trainingsdaten

```python
# Beste Pipeline mit der Kreuzvalidierung trainieren und Ausgabe der Perfomanz auf den Trainingsdaten
final_pipelines = []
for piepline in best_params:
    train_labels = cross_val_predict(piepline, X_train, y_train, cv=10)
    # Precision/Recall/F-Wert berechnen
    final_pipelines.append(piepline)
    print(classification_report(y_train, train_labels))
```

```python
# Jetzt den Lerner ein letztes Mal auf allen Trainingsdaten trainieren und dann auf den Testdaten evaluieren

# Lerner auf den gesamten Trainingsdaten trainieren
for final_pipeline in final_pipelines:
    final_pipeline.fit(X_train, y_train)
    print("Default-Score des Klassifizierers: Accuracy=",final_pipeline.score(X_dev, y_dev), "\n")
    # Labels vorhersagen lassen und dann Precision/Recall/F-Wert berechnen
    test_labels = final_pipeline.predict(X_dev)
    print(classification_report(y_dev, test_labels))
```

### Anforderung 4
Für unseren Datensatz hat der XGBoost Klassifizierer mit einer Accuracy von 0,7264 am besten abgeschnitten. Mit Abstand am schlechtesten war der Naive Bayes Klassifizierer mit einer Accuracy von 0,6124. Der letzte Klassifikations-Report in der Ausgabe ist für unseren besten Lerner den XGBoost Klassifizierer. Mit einem F-Score von 0,73 ist dieser allgemein um 1% besser bei der Vorhersage von erfolgreichen Projekten (Klasse 1). Dies liegt daran, dass der Recall für die unerfolgreichen Projekte (Klasse 1) mit 0,69 deutlich unter dem Recall von 0,76 für die erfolgreichen Projekte (Klasse 1) liegt. Das bedeutet unser Lerner sagt sehr häufig die Klasse 1 (erfolgreichen Projekte) voraus, wodurch sich ein guter Recall für diese Klasse ergibt. Auf der anderen Seite ist daher der Recall für Klasse 0 (unerfolgreichen Projekte) schlecht, wodurch weniger tatsächlich fehgeschlagene Projekte (Klasse 0) richtig vorhergesagt wurden. Im Umkehrschluss ist dafür die Vorhersage für ein erfolgreiches Projekt (Klasse 1) nicht so zuverlässig wie die Vorhersage für ein fehlgeschlagenes Projekt (Klasse 0). Da es uns, aber wichtiger ist vorhersagen zu können, ob ein Projekt erfolgreich sein wird, ist es für uns besser, wenn die Klasse 1 (erfolgreichen Projekte) den besseren Recall hat, so lange die Precision, wie hier nicht deutlich schlechter ist. Unser Ergebnis für unser Klassifikationsmodell könnte durch die Ermittlung verschiedener Feature und einem umfangreicheren Datensatz vermutlich noch deutlich verbessert werden, aber es ist unserer Meinung nach für einen ersten Versuch schon zufriedenstellend. Für einen Einsatz in der Praxis müsste das Ergebnis noch verbessert werden.
