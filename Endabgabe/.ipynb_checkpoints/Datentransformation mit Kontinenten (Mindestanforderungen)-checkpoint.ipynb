```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os
from os.path import dirname
import datetime as dt
from random import randint
import pycountry_convert as pc

notebook_path = os.path.abspath("Datentransformation mit Kontinenten (Mindestanforderungen).ipynb")
csv_path = os.path.join(os.path.dirname(dirname(notebook_path)), "data/ks-project-edited-numpy-pandas.csv")
```

```python
df = pd.read_csv (csv_path, low_memory=False)
df.head()
```

```python
# Entfernen von ungültigen Einträgen
index = df.index
condition = (df["country"] == 'N,0"') | (df["state"] == "undefinded")
df_filtered_indices = index[condition]
df_clean = df.drop(index=df_filtered_indices)
```

```python
# Zweites Dataframe für den Merge erzeugen mit den Projekt IDs
df_continents = pd.DataFrame(df_clean['ID'])
df_continents
```

```python
# Liste mit den passenden Kontinenten füllen zum ersten Dataframe
continents = []
for index, row in df_clean.iterrows():
    continent_name = pc.country_alpha2_to_continent_code(row['country'])
    continents.append(continent_name)
```

```python
# Zweites Dataframe um das Feature "continent" ergänzen
df_continents["continent"] =  continents
df_continents
```

### Anforderung 1

```python
# Mergen der beiden Dataframes mit der ProjektID
df_country_continent = df_clean.merge(df_continents, on ='ID')
df_country_continent
```

### Anforderung 2

```python
# Bilden einer Pivot-Tabelle mit den Ländern als Spalten und den Kontinenten als Zeilen-Labeln
# Zählen wie viele Projekte vorkommen
df_pivot= df_country_continent.pivot_table(index='continent', columns='country', values='ID', aggfunc='count', fill_value=0.0)
df_pivot
```

### Anforderung 3

```python
# Ermitteln wie viel Geld gesammelt wurden in den verschienden Ländern und Kontinenten
df_pivot= df_country_continent.pivot_table(index='continent', columns='country', values='usd_pledged_real', aggfunc='sum', margins = True, fill_value=0.0)
df_pivot
```
