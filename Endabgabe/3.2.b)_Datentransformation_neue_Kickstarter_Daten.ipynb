### In diesem Notebook haben wir unseren bestehenden Datensatz, um das Feature CreatorID aus den neuen Kickstarter Daten erg√§nzt

## Datensatz einlesen

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
from os.path import dirname
import datetime as dt
import seaborn as sns

notebook_path = os.path.abspath("3.2.b)_Datentransformation_neue_Kickstarter_Daten.ipynb")
csv_path_creator_id = os.path.join(os.path.dirname(dirname(notebook_path)), "data/ks-project-edited-creator-id.csv")
csv_path_normal = os.path.join(os.path.dirname(dirname(notebook_path)), "data/ks-project-edited-numpy-pandas.csv")
```

```python
df_creator_id = pd.read_csv (csv_path_creator_id, low_memory=False)
df_normal = pd.read_csv (csv_path_normal, low_memory=False)

# Dataframe mit der ProjektID und der creator id
df_creator_id = df_creator_id.drop(columns="Unnamed: 0")
df_creator_id.head()
```

```python
# Unser originaler Datensatz
df_normal.head()
```

```python
# Mergen der beiden Dataframes
df_merged = df_normal.merge(df_creator_id, on='ID')
df_merged.head()
```

```python
# Nun haben wir nur noch 129580 Eintr√§ge. Davor hatten wir ca. 370000
df_merged.info()
```

```python
df_grouped_projectsPerCreaor = df_merged.groupby(['creator_id']).size().to_frame('numberOfProjects').reset_index().sort_values("numberOfProjects", ascending=False)
df_grouped_countPerProjectNumber = df_grouped_projectsPerCreaor.groupby(['numberOfProjects']).size().to_frame('count').reset_index().sort_values("count", ascending=False)
```

```python
# Erstellen von Dataframe mit Ceratorn, welche geanu 1 Projekt bzw mehrere Projekte erstellt haben
df_withOneProject = df_grouped_projectsPerCreaor[df_grouped_projectsPerCreaor['numberOfProjects'] ==1]
df_withMultipleProjects =  df_grouped_projectsPerCreaor[df_grouped_projectsPerCreaor['numberOfProjects'] > 1]
```

```python
# Balkendiagram, das zeigt wie h√§ufig Creator Projekte erstellen
fig, ax = plt.subplots(figsize=(10,6))
ax = sns.barplot(data = df_grouped_countPerProjectNumber, x = "numberOfProjects", y="count")
plt.title("Anzahl der Projekte die ein Creator erstellt", pad=20,fontweight="bold")
plt.ylabel("H√§ufigkeit")
plt.xlabel("Anzahl der Projekte eines Creators")
plt.show()
```

```python
# Es gibt bei diesem zusammengef√ºgten Datensatz mehr erfolgreiche Projekte als  fehlgeschlagene Projekte 
# Dies entspricht nicht der Verteilung des eigentlichen Datensatzes, 
# da dort mehr fehlgeschlagene als erfolgreiche Projekte sind (ca. nur 1/3 der Projekte war erfolgreich)
fig, axs = plt.subplots(ncols=2, figsize=(15,15))
colors = sns.color_palette('pastel')[0:6]

def my_autopct(pct):
    return ('%.2f%%' % pct) if pct >= 5 else ''
#create pie chart
axs[0].pie(df_merged['state'].value_counts(), labels= df_merged['state'].value_counts().index.tolist(),colors = colors, autopct=my_autopct)
axs[1].pie(df_normal['state'].value_counts(), labels= df_normal['state'].value_counts().index.tolist(),colors = colors, autopct=my_autopct)

axs[0].set_title("Erfolg oder Misserfolg der Projekte (Zusammengef√ºgter Datensatz)", fontweight="bold", fontsize=12, pad=10)
axs[1].set_title("Erfolg oder Misserfolg der Projekte (Originaler Datensatz)", fontweight="bold", fontsize=12, pad=10)

plt.tight_layout()        
plt.show()
```

```python
# Ermitteln der ersten Projekte eines Creators
df_merged_sorted = df_merged.sort_values(by=["creator_id","launched"])
df_creators_first_project = df_merged_sorted.groupby("creator_id").first().reset_index()
```

```python
# Ermitteln aller Creator mit meheren Projekten
df_creators_multiple_projects = df_merged_sorted[df_merged_sorted.duplicated(subset=["creator_id"],keep=False)]
# Entfernen des ersten Projekts eines Creators
df_creators_multiple_projects = df_creators_multiple_projects[df_creators_multiple_projects.creator_id.duplicated()]
```

```python
# Projekte die Creator nach ihrem ersten Projekt starten sind viel h√§ufiger erfolgreich
fig, axs = plt.subplots(ncols=2, figsize=(15,15))
colors = sns.color_palette('pastel')[0:6]

def my_autopct(pct):
    return ('%.2f%%' % pct) if pct >= 5 else ''
#create pie chart
axs[0].pie(df_creators_first_project['state'].value_counts(), labels= df_creators_first_project['state'].value_counts().index.tolist(),colors = colors, autopct=my_autopct)
axs[1].pie(df_creators_multiple_projects['state'].value_counts(), labels= df_creators_multiple_projects['state'].value_counts().index.tolist(),colors = colors, autopct=my_autopct)

axs[0].set_title("Erfolg oder Misserfolg der Projekte (Erstes Projekt eines Creators)", fontweight="bold", fontsize=12, pad=10)
axs[1].set_title("Erfolg oder Misserfolg der Projekte (Nach dem ersten Projekt)", fontweight="bold", fontsize=12, pad=10)

plt.tight_layout()        
plt.show()
```

```python
# Ermitteln der erfolgreichen ersten Projekte eines Creators
df_creators_first_project_succ = df_creators_first_project[df_creators_first_project["state"] == "successful"]
# Ermitteln der Creator, welche mehrere Projekte haben und deren erstes Projekt erfolgreich ist
df_creators_first_project_succ_mult = df_creators_multiple_projects[df_creators_multiple_projects["creator_id"].isin(df_creators_first_project_succ["creator_id"])]
# Ermitteln des zweiten Projects eines Creators, wenn sein erstes erfolgreich ist
df_creators_second_project_succ = df_creators_first_project_succ_mult.groupby("creator_id").first().reset_index()
```

```python
# Ermitteln der fehlgeschlagenen ersten Projekte eines Creators
df_creators_first_project_fail = df_creators_first_project[df_creators_first_project["state"] == "failed"]
# Ermitteln der Creator, welche mehrere Projekte haben und deren erstes Projekt fehlgeschlagen ist
df_creators_first_project_fail_mult = df_creators_multiple_projects[df_creators_multiple_projects["creator_id"].isin(df_creators_first_project_fail["creator_id"])]
# Ermitteln des zweiten Projects eines Creators, wenn sein erstes fehlgeschlagen ist
df_creators_second_project_fail = df_creators_first_project_fail_mult.groupby("creator_id").first().reset_index()
```

```python
fig, axs = plt.subplots(ncols=2, figsize=(15,9))
ax1 = sns.countplot(x='state', data=df_creators_second_project_succ,ax=axs[0], order = df_creators_second_project_succ['state'].value_counts().index)
ax1.set_title('Wahrscheinlichkeit, dass das zweite Projekt erfolgreich ist,\nwenn das erste erfolgreich war', pad=20, fontweight="bold")
ax1.set_xlabel('Erfolg oder Misserfolg')
ax1.set_ylabel('Anzahl der Projekte')

ax2 = sns.countplot(x='state', data=df_creators_second_project_fail,ax=axs[1], order = df_creators_second_project_fail['state'].value_counts().index)
ax2.set_title('Wahrscheinlichkeit, dass das zweite Projekt erfolgreich ist,\nwenn das erste nicht erfolgreich war', pad=20, fontweight="bold")
ax2.set_xlabel('Erfolg oder Misserfolg')
ax2.set_ylabel('Anzahl der Projekte')
        
plt.tight_layout(pad=10.0)        
plt.show()
```
