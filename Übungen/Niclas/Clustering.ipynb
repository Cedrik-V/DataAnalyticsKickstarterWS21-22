```python
# Imports

from sklearn import cluster, datasets

import matplotlib.pyplot as plt

from sklearn import metrics
```

```python
# Daten herunterladen
faces = datasets.fetch_olivetti_faces()
```

```python
# Wir können uns die einzelnen Bilder auch anschauen:

# Bildgröße bestimmen    
plt.figure(figsize=(24, 24)) 

# Alle Bilder in einen Plot mit 40 Reihen und 10 Spalten sammeln, dann anzeigen 
for i in range(0,len(faces["images"])):
    plt.subplot(40,10,i+1)
    plt.axis("off")
    plt.imshow(faces["images"][i])
    
# ein einzelnes Bild anschauen

plt.imshow(faces["images"][20])
```

```python
# K-Means anwenden. Wir wissen, dass 40 Zielklassen existieren, daher erwarten wir n_clusters=40 Cluster
# Bei Daten ohne Zielklasse muss dieser Parameter variiert werden!

# Wir geben an, mit welchem random_state die zufälligen Anfangszentroide festgelegt werden sollen.

kmeans = cluster.KMeans(n_clusters=40, random_state=3).fit(faces["data"])
```

```python
# Welche Cluster wurden für die Instanzen gefunden?
print(kmeans.labels_)
```

```python
# Wie groß sind die Cluster? Wir möchten 40 Cluster mit je 10 Datenpunkten sehen.

for i in range(0,40):
    print("Name:", i, "Größe: ", list(kmeans.labels_).count(i))
  
  
# aber diese sind hochdimensional und daher intuitiv nicht sehr informativ.
# Sie entsprechen auch nicht notwendigerweise einem echten Bild.
```

```python
hom = metrics.homogeneity_score(faces["target"],kmeans.labels_)
com =  metrics.completeness_score(faces["target"],kmeans.labels_)
vmeasure = metrics.v_measure_score(faces["target"],kmeans.labels_)

print(hom)
print(com)
print(vmeasure)
```

```python
inertia = kmeans.inertia_
sil = metrics.silhouette_score(faces["data"],kmeans.labels_)

print(inertia)
print(sil)
```

```python
def bench_k_means(kmeans, name, data, labels):
    """Benchmark to evaluate the KMeans initialization methods.

    Parameters
    ----------
    kmeans : KMeans instance
        A :class:`~sklearn.cluster.KMeans` instance that has already been trained
    name : str
        Name given to the strategy. It will be used to show the results in a
        table.
    data : ndarray of shape (n_samples, n_features)
        The data whose clusters should be evaluated.
    labels : ndarray of shape (n_samples,)
        The labels used to compute the supervised clustering metrics.
    """
    
    # Inertia
    results = [name, kmeans.inertia_]

    # The silhouette score requires the full dataset
    results += [
        metrics.silhouette_score(data, kmeans.labels_,
                                 metric="euclidean")
    ]
    
    # Supervised metrics which require the true labels and cluster
    # labels
    clustering_metrics = [
        metrics.homogeneity_score,
        metrics.completeness_score,
        metrics.v_measure_score,
    ]
    results += [m(labels, kmeans.labels_) for m in clustering_metrics]

    # Show the results
    formatter_result = ("{:9s}\t{:.0f}\t{:.3f}\t{:.3f}\t{:.3f}\t{:.3f}")
    print(formatter_result.format(*results))
```

```python
# Aufrufen der Benchmark mit unserem trainierten kmeans und einer zufälligen Zuweisung (zum Vergleich)
# Fügen Sie hier auch Ihren günstigeren random_state von oben ein!

print(82 * '_')
print('init\t\tinertia\tsil\thom\tcom\tv-meas')

# unser kmeans-Modell (zur Sicherheit neu trainieren)
kmeans = cluster.SpectralClustering(n_clusters=40).fit(faces["data"])
bench_k_means(kmeans=kmeans, name="k-means_3", data=faces["data"], labels=faces["target"])

# unser kmeans-Modell (random_state 42)
kmeans = cluster.KMeans(n_clusters=40, random_state=42).fit(faces["data"])
bench_k_means(kmeans=kmeans, name="k-means_42", data=faces["data"], labels=faces["target"])

# unser kmeans-Modell (10x trainiert; das Modell mit der besten inertia gewinnt)
kmeans = cluster.KMeans(n_clusters=40, n_init=10).fit(faces["data"])
bench_k_means(kmeans=kmeans, name="k-means_opti", data=faces["data"], labels=faces["target"])

# zufällige Cluster-Zuordnung erzeugen
kmeans_rand = cluster.KMeans(init="random", n_clusters=40, random_state=0).fit(faces["data"])
bench_k_means(kmeans=kmeans_rand, name="random", data=faces["data"], labels=faces["target"])

print(82 * '_')
```
